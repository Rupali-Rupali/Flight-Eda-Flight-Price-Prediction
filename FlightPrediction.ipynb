## Flight Price Prediction(EDA + Feature Engineering)

Flight price prediction is an intriguing field that combines data analysis, machine learning, and domain knowledge to estimate the cost of air travel. Let’s delve into the theories and techniques behind predicting flight prices:

Factors Impacting Flight Prices:

Various factors influence flight prices, including:
Distance: Longer flights tend to be more expensive.
Flight Time: The time of day affects prices; peak hours may cost more.
Number of Stops: Direct flights are often pricier than those with layovers.
Airline: Different carriers have varying pricing strategies.
These factors create patterns that machine learning models can learn from to predict future prices.

### Dataset Information:
Key columns in the flight fare prediction dataset include:

    *  Airline: Types of airlines (e.g., Indigo, Jet Airways).
    *  Date_of_Journey: Departure date.
    *  Source and Destination: Travel origin and destination.
    *  Route: Flight path.
    *  Arrival_Time: Arrival time.
    *  Duration: Flight duration.
    *  Total_Stops: Number of layovers.
    *  Additional_Info: Details about amenities.
    *  Price: Total fare.

## Objective:

By employing machine learning algorithms, particularly regression techniques, we aim to predict flight ticket prices accurately.

### Extracting Date,Year, Month from Date of Journey feature

```python
final_df['Date_of_Journey'].str.split('/')
```

```python
##Fetaure Engineering Process
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
# Changing the datatype of Date, Month , Year
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(5)
```

### Extracting Arrival_hour, Arrival_min from Arrival_Time feature

```python
final_df['Arrival_Time'].str.split(' ')
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

```python
# Converting datatype from object to int
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
# Dropping Arriving Time from the dataset
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(5)
```

### Extracting Dept_hour, Dept_min from Dept_Time feature

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```
